```python
def Convert_real (HH, dr):
    #for i range(len(r)):
    pos= np.matmul(HH, dr)
    return pos
    
def Convert_scale(HHi, dr):
    pos = np.matmul(HHi, dr)
    return pos
    
def PBC_condition(rr):
    nlen = len(rr)
    #print (nlen)
    assert nlen == 3
    
    if (rr[1] < -0.5): rr[1] = rr[1] + 1
    if (rr[1] >= 0.5): rr[1] = rr[1] - 1
    if (rr[2] < -0.5): rr[2] = rr[2] + 1
    if (rr[2] >= 0.5): rr[2] = rr[2] - 1
    if (rr[0] < -0.5): rr[0] = rr[0] + 1
    if (rr[0] >= 0.5): rr[0] = rr[0] - 1
    return rr;
    
def Cutoff(ntype):
    n= (len(ntype))
    cutoff_val= np.empty([n, n])
    for i in range(n):
        for j in range(n):
            cutoff_val[i][j]= 1.5
    cutoff_val[1][1] = 1.2
    cutoff_val[1][0] = 1.4
    cutoff_val[0][1] = 1.4
    cutoff_val[1][3] = 1.2
    cutoff_val[3][1] = 1.2
    cutoff_val[1][2] = 1.2
    cutoff_val[2][1] = 1.2
    cutoff_val[2][2] = 1.2
    cutoff_val[2][1] = 1.2
    return cutoff_val
```
